```python
X = df.drop(['MadrsMean'], axis=1)
y = df['MadrsMean']
```

## Split Data 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state = 42)
```

KNeighbors Regressor

```python
knn_model = KNeighborsRegressor()
```

```python
model = knn_model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Test Error 

```python
# RMSE
mean_squared_error(y_test, y_pred)
```

```python
#RMSLE
np.sqrt(mean_squared_error(y_test, y_pred))
```

## Train Error 

```python
mean_squared_error(y_train, model.predict(X_train))
```

```python
np.sqrt(mean_squared_error(y_train, model.predict(X_train)))
```

## Train error different k values (not validated) 

```python
for k in range(18):
    k=k+1
    k_model = KNeighborsRegressor(n_neighbors = k).fit(X_train, y_train)
    y_pred = k_model.predict(X_train)
    error = np.sqrt(mean_squared_error(y_train, y_pred))
    print('k value = ' , k , 'error value =' , error)
```

## Test error different k values (not validated)


```python
for k in range(18):
    k=k+1
    k_model = KNeighborsRegressor(n_neighbors = k).fit(X_train, y_train)
    y_pred = k_model.predict(X_test)
    error = np.sqrt(mean_squared_error(y_test, y_pred))
    print('k value = ' , k , 'error value =' , error)
```
